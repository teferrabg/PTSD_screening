```python
import pandas as pd

# Load dataset
df = pd.read_csv("/content/detailed_lables.csv")

# Keep only the required columns
columns_to_keep = [
    "Participant",
    "PCL-C_1_Memories", "PCL-C_2_Dreams", "PCL-C_3_Reliving", "PCL-C_4_Upset",
    "PCL-C_5_Physical", "PCL-C_6_ThoughtAvoidance", "PCL-C_7_ActivityAvoidance",
    "PCL-C_8_TroubleRemembering", "PCL-C_9_NoInterest", "PCL-C_10_FeelDistant",
    "PCL-C_11_FeelNumb", "PCL-C_12_Future_Cut_Short", "PCL-C_13_Sleep",
    "PCL-C_14_Irritability", "PCL-C_15_Concentration", "PCL-C_16_HyperAlert",
    "PCL-C_17_Jumpy", "PTSD_severity", "gender"
]

df_filtered = df[columns_to_keep]

# Keep only rows where Participant is between 300 and 400
df_filtered = df_filtered[(df_filtered["Participant"] >= 300) & (df_filtered["Participant"] <= 400)]

# Save new dataset
output_path = "/content/detailed_labels_filtered.csv"
df_filtered.to_csv(output_path, index=False)

print(f"Filtered dataset saved to: {output_path}")
```
